```python
# importing pandas library as object name pd
import pandas as pd

# With the help of pd we will access panda library functions
```

```python
# function -> Series()

# Series - Series is something same as 1-d array, it contains non-numeric data, like date, time, boolean etc.
# Series are part of data frames. 

s1 = pd.Series([1,2,5])
s1
```

```python
# function -> DataFrame()

# DataFrame - DF is a table with rows and collumns. Their rows have index and columns have meaningful names. 
# DF can be build from dictionaries, Json object, readings csv files etc.

df_dict = pd.DataFrame({
    'rollno' :[1,2,3,4,5],
    'name':['t','o','s','i','f'],
    'branch':[10,11,12,13,14]
})

# for accesing data frame we can simply call it's object
df_dict
```

```python
# function -> read_csv()

# Now the most interesting part of pandas library
# accesing data from a csv file
# First we need to keep the file location inside read_csv function
#              
#         example  - df_csv = read_csv(r'F:\Machine Learning\Pandas operations\machine_learning\basics')       
#                       ^                                            ^
#                    by this we                                   location of
#                  are storing our                                our csv file
#               DF into df_csv object
#
df_csv = pd.read_csv('housing.csv')
```

```python
# function -> head()

# Some basic functions which can applied on the dataFrame 
# to see its few rows and columns from top or bottom 

# To see top-most rows head() function is used.
# It takes (5) as its default parameter(if left empty) or we can pass our parameter like 8,10,20,etc.

df_csv.head()
```

```python
# function -> tail()

# To see bottom rows tail() function is used.
# It takes (5) as its default parameter(if left empty) or we can pass our parameter like 8,10,20,etc.

df_csv.tail()
```

```python
# function -> info()

# To see basic information of DF, info() function is used which tells information about
# columns of a DF its Datatypes, Non-Null count [if it contains any null values], 
# memory usage of DF, RangeIndex etc.

df_csv.info(verbose= True)

# it has a parameter named verbose which is true by default 
# it shows all columns details.
```

```python
# function -> describe()

# It shows some mathematical operations on all integers column. 
# such as count, mean, std, min, 25%, 50%, 75%, max.

df_csv.describe()
```

```python
# shape

# It shows shape i.e., no. rows and columns of a dataFrame.
df_csv.shape
```

```python
# columns

# It shows all columns name in form of Series.
df_csv.columns
```

```python
# values 

# It converts a Dataframe into an array array format now 
# It becomes same as arrays in numpy.

df_csv.values
```

```python
# For accessing columns

# If we want to access a particular column of DF than we can simply use column name.
# And access the column. Our output is in format of Series.

# First Method
df_csv.bedrooms
# Second Method
df_csv['bedrooms']


```

```python
# If we want to access more than one column from a DF than 
# we pass a list inside index of our DF object

# First Method
c = df_csv[   ['bedrooms','price']  ]

# Second Method
a = ['bedrooms','price']
c = df_csv[a]
c
```

```python
# For accessing rows
# Indexing done for row

df_csv[2:3]
#      ^ ^
#  start Stop
```

```python
# function -> sort_values()

# this function can sort DF according to passed parameters which is column name
# It has two parameters (by='col_name', ascending=False)
#                                ^             ^
#                        by which col       True by
#                      it is to be sort     default

df_csv.sort_values(by='price',ascending=False)
```

```python
# function -> apply()

# apply() function iterates whole column as series and 
# lamda function performs operation and returns value to apply() and apply() stores it into data frame.

df_csv.parking = df_csv.parking.apply(lambda x : x+1)

df_csv
```

```python
# iloc 
# it takes meaningful columns names as index in integer form.

df_csv.iloc[1:3, 2:4]
#        ^
# it also can be said "slicing without labels"
```

```python
#loc
# it takes meaningful columns as index with their real names.

df_csv.loc[1:3, ['bedrooms', 'bathrooms']]
#       ^
# it also can be said "slicing with labels"

# Here we can see Outputs in both cases are same 
# but in first case it uses columns indexing as integer format
# and in second case it uses columns indexings as their real names.
```

```python
# More Operations With "iloc".

df_csv.iloc[[1,2,5],]
#              ^
# Here we pass a list of numbers which are index of rows to be fetched
# Now in deep we can understand that list [1,2,5] is passed in first place 
# before coma in iloc[ here_is_list , ]
# Now list provide list_item one by one to iloc[list_item] than iloc fetch
# these indexes rows.
```

```python
# More operations with "iloc".

df_csv.iloc[[x for x in range(10) if x%2 == 0],]
#                   ^
# Here we are using a comprension list for creating a list of even numbers
# these even numbers are stored in list, each number obtained in for loop interations
# is given to x and this x append it into list. This is how its working
```

```python
# logical operations with DataFrames

# Here we are using logical & operator 
# We got three cases here. whether

df_csv[(df_csv.bedrooms>4) & (df_csv.guestroom == 'yes') & (df_csv.basement == 'yes')] 
#               ^                          ^                             ^
#       house whose bedrooms      house which has                  house which has
#         are more than 4            guestroom                       basement
#  
# In output we got a dataframe where all these 3 conditions are satisfied
# And We can also use logical or (|) operator when we need at least one condition is to be satisfied.
```

```python
# function -> unique()

# With the help of this function we can find the unique values of a particular column
# simply we can use it as,

df_csv.mainroad.unique()
#             ^
#        column_name
#     for which we need
#       unique values
```

```python
# With the help of unique() function next function becomes more easy and useful

# function -> map()

# it takes a dictionary as parameter by which it change the items in columns.

df_csv.mainroad = df_csv['mainroad'].map({
    'yes':1,
    'no':0
})

# Here we are using this dictionary to change yes -> 1 and no -> 0
# It is also helpful as we can't perform mathematical operations on strings 

df_csv[0:3]         
```

```python
# function -> groupby()

# It is used to create specific groups according to the given columns values. 
# We can create groups using single columns or multiple columns

df_csv.groupby(['mainroad']).sum()
```

```python
# function -> getgroup()

# This function is used to get the specific group of dataframe according to passed key.

df_csv.groupby(['mainroad']).get_group(1)
#                                      ^
#                              here in our case 
#                              it can be 0 or 1
```

```python
# function -> groups()

# It returns data dictionary of all rows in group format.

obj = df_csv.groupby(['mainroad'])
obj.groups
```

```python
# Creating new columns 

# Suppose if we want to create a new column which contains the value of ratio between 
# bedrooms and bathrooms than we can simply do,

df_csv['bed_bath_ratio'] = df_csv.bedrooms / df_csv.bathrooms
#             ^
#      This will contain the name of new column

df_csv.iloc[0:4, 12:14]
```

```python
# function -> aggr()

df_csv['area'].agg('mean')
#        ^            ^
#     col_name | mean, median or mode
```

```python
# function -> pivot_table()

# It is an alternative of group_by function but simple than it.
# It creates group of given columns with all values.

df_csv.pivot_table(index=['mainroad'],
                    values=['price','area'],
                    aggfunc='median',
                    columns='basement'
)

# pivot_table provides functionality to create aggregate tables

# * aggfunc -> sum, mean, median etc.
# * index   -> list of columns used to group the data.
# * values  -> columns on which aggfunc will be applied.
# * columns -> parameter which takes list of columns we want in pivot_table.
```
